### Does any column contain alpha and numeric data? Decide how to clean it.

```python
categorical_df # Customer, Policy and Renew Offer Type have alphanumerical

categorical_df = categorical_df.drop(['Customer'], axis = 1) # Dropping this coloumns
```

```python
categorical_df
```

```python
categorical_df['Policy'].value_counts() # We are going to change 3-C, 2-B, 1-A
categorical_df['Renew Offer Type'].value_counts() # We are going to chanfe 1-A, 2-B, 3-C, 4-D

def clean_data(df):

# Change alphanumeric Policy
    df['Policy'] = [x.replace('1','A') for x in df['Policy']]
    df['Policy'] = [x.replace('2','B') for x in df['Policy']]
    df['Policy'] = [x.replace('3','C') for x in df['Policy']]

# Change alphanumeric Policy
    df['Renew Offer Type'] = [x.replace('1','A') for x in df['Renew Offer Type']]
    df['Renew Offer Type'] = [x.replace('2','B') for x in df['Renew Offer Type']]
    df['Renew Offer Type'] = [x.replace('3','C') for x in df['Renew Offer Type']]
    df['Renew Offer Type'] = [x.replace('4','D') for x in df['Renew Offer Type']]

# Change to snakecase
    df.columns =[colname.lower().replace(' ','_') for colname in df.columns]
    return df
```

```python
categorical_df = clean_data(categorical_df)

categorical_df
```
